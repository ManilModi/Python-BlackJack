```python

import random

values={'Two':2, 'Three':3, 'Four':4, 'Five':5, 'Six':6, 'Seven':7, 'Eight':8,
        'Nine':9, 'Ten':10, 'Jack':10, 'Queen':10, 'King':10, 'Ace':11}

suits=("Hearts", "Diamonds", "Spades", "Clubs")
ranks=("Two", "Three", "Four", "Five", "Six", "Seven", "Eight", "Nine", "Ten", "Jack", "Queen", "King", "Ace")

status=True

class Card:

    def __init__(self,suit,rank):
        self.suit = suit
        self.rank = rank
        
    def __str__(self):
        return self.rank + ' of ' + self.suit
```

```python
class Deck:
    
    def __init__(self):
        self.deck = []  
        for suit in suits:
            for rank in ranks:
                self.deck.append(Card(suit,rank))  # build Card objects and add them to the list
    
    def __str__(self):
        deck_comp = ''  # start with an empty string
        for card in self.deck:
            deck_comp += '\n '+card.__str__() # add each Card object's print string
        return 'The deck has:' + deck_comp

    def shuffle(self):
        random.shuffle(self.deck)
        
    def deal(self):
        single_card = self.deck.pop()
        return single_card
```

```python
class Hand:
    
    def __init__(self):
        self.card=[]
        self.value=0
        self.aces=0
        
    def add_cards(self,new_cards):
        self.card.append(new_cards)
        self.value+= values[new_cards.rank]
        if new_cards.rank == 'Ace':
            self.aces += 1 
        
    def if_Ace(self):
        
        while self.value>21 and self.aces:
            self.value-=10
            self.aces-=1
        
    
```

```python
class Chips:
    
    def __init__(self):
        self.total=100
        self.bet=0
        
    def win_bet(self):
        self.total+=self.bet
    
    def lose_bet(self):
        self.total-=self.bet
```

```python
def have_bet(chips):
    
    while True:
        
        try:
            chips.bet=int(input("How many chips do you want to bet? \n"))
        except:
            print("Sorry, a bet must be an integer.")
        finally:
            if chips.bet>chips.total:
                print("Sorry, you do not have sufficient chips to bet!\n")
            else:
                break
    
```

```python
def hit(deck,hand):
    hand.add_cards(deck.deal())
    hand.if_Ace()
```

```python
def take_hits(deck,hand):
    
    global status
    
    while True:
        x=input("Would you like to hit or stand? Enter h or s: \n")
    
        if x[0].lower()=='h':
            hand.add_cards(deck.deal())
            hand.if_Ace()
        elif x[0].lower()=='s':
            print("The player wants to stand.  Dealer can play now:\n")
            status=False
            break
            
        else:
            print("Please try again!")
            continue
```
